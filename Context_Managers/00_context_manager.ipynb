# **Class** Context Managers
- Son definidos mediante 1 clase con 2 methods necesarios y 1 opcional que requieren e implementan el siguiente orden:
    - `def __init__(self)` *(Opcional)*:
        - sirve para inicializar el recurso, por ejemplo guardar nombres u otros metadatos 
    - `def __enter__(self)` (**Obligatorio**):
        - sirve para realizar acciones dentro del bloque `with`
        - realiza las acciones de crear, modififcar, acceder, iniciar una conexión
        - **retorna valor asignado en**  la implementacion del Context Manager (frontend del Context Manager) a **la variable** *asignada con `as` usando `yield`* dentro de este metodo `__enter__()`, ejemplo:
            ```python
            with open('ejemplo.txt', 'w+r') as file:
                print(
                    type(file)
                )
            
            ```
    - `def __exit__(self, exc_type, exc_value, traceback)` (**Obligatorio**):
        - libera o cierra recursos (conexiones de red, cerrar archivos abiertos o bases de datos)
        - pueden ayudarnos a manejar errores 
        - podemos eligir evitar lanzar errores retornando `False`, o `True` para retornar una **excepción** / **exception**
            ```python
            def __exit__(self, exc_type, exc_value, traceback):
                return True     #-> No retorna exceptions si existe alguna dentro de 'with'
                #return False   #-> retorna exceptions si existe alguna dentro de 'with'
            ```
        - > [!NOTE] Requiere 4 argumentos obligatorios:
            > - `self` - referencia al objeto creado
            > - `exc_type` - Tipo de excepción que fue lanzada
            > - `exc_value` - Valor de la excepción en caso de ser proporcionado
            > - `traceback` - Objeto con más información de la excepción

````python
#ejemplo 1.0
class CntxMngr_FileCreator:
    def __init__(self, nombre):
        self.file_name = nombre

    def __enter__(self):
        self.file_resource = open(self.file_name, 'w')

    def __exit__(self, exec_type, exec_value, traceback):
        self.file_resource.close()
        return True



with CntxMngr_FileCreator("ejemplo.txt"):
    raise Exception


# Abstracción de clase para leer y añadir al final del archivo si ya existe
class Cnxtxt_FileModifier:
    def __init__(self, file_name, file_content):
        self.file_name = file_name
        self.file_content = file_content

    def __enter__(self):
        self.file_resource = open(self.file_name, 'a')

        self.file_resource.write(f"\n{self.file_content}")

    def __exit__(self, exc_type, exc_value, traceback):
        return False # retorna detalles del error si encuentra alguno
````

````python
#ejemplo 1.1

# ejecución concurrente (haciendo y parando varias tareas pa una ejecución "paralela")
import threading

#candado de acceso a un recurso si multiples archviso necesitan acceder a el 
lock = threading.Lock()

#1 definir funcion / tarea a realizar por los locks de manera concurrente
    # implementar Clase Cnxtxt_FileModifier para:
        # - leer archivo
        # - agregar al final
def agregar_contenido(archivo, contenido ):
    with lock:
        with Cnxtxt_FileModifier( archivo, contenido):
            print("contenido añadido")
    print("código que accede a un recurso")

#2 usar bloque with lock para restringir el acceso del archvio de forma concurrente "paralela" si algun lock lo abrio primero y aun no termina de usarlo
hilo_1 = threading.Thread(target=agregar_contenido, args=("ejemplo.txt", "hilo_1"))
hilo_2 = threading.Thread(
    target=agregar_contenido, args=("ejemplo.txt" , "hilo_2")
)

hilo_1.start()
hilo_2.start()
hilo_2.join()
hilo_1.join()
````
